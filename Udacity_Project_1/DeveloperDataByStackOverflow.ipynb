## Evaluating Results and Business Understanding

### Question 1
### What is the average salary of particular UnderGrad Major and what is the average salary of particular Formal Education?


#### Average salary of UnderGrad Major and Average salary according to Formal Education

```python
#Number of values of a particular major
data.MajorUndergrad.value_counts()
```

```python
#Undergrad major avg salary
MajorAvgSalary=data.groupby('MajorUndergrad')['Salary'].mean().sort_values(ascending=False)
MajorAvgSalary
```

```python
# Plot a horizontal bar chart from the MajorAvgSalary Series object
plt.barh(MajorAvgSalary.index, MajorAvgSalary.values)
plt.xlabel('Average Salary')  # Average Salary label
plt.ylabel('Undergraduate Major')  # Undergraduate Major label
plt.title('Average Salary by Undergraduate Major')  # Title of the chart
plt.show()
```

```python
#Number of value of particular Formal Education
data['FormalEducation'].value_counts()
```

```python
#Particular Formal Education  average salary
FormalAvgSalary=data.groupby('FormalEducation')['Salary'].mean().sort_values(ascending=False)
FormalAvgSalary
```

```python
# Plotting the graph horizontally 
plt.barh(FormalAvgSalary.index, FormalAvgSalary.values)
plt.xlabel('Average Salary')  # X-axis label
plt.ylabel('Formal Education Level')  # Y-axis label
plt.title('Average Salary by Formal Education Level')  # Title of the chart
plt.show()
```

##### Result :
Psychology  have the highest average salary among Majors in undergrad.

Doctoral degree is the highest average salary among Formal Education.

## Question 2
### Which company size and company type have highest salary?

```python
#Number of company of particular company size
data['CompanySize'].value_counts()
```

```python
#Number of companyType
data['CompanyType'].value_counts()
```

```python
#Avg salary of company size and company type
data.groupby(['CompanySize','CompanyType'])['Salary'].mean().sort_values(ascending=False)
```

```python
#Grouping Company Type and Company size and finding average salary in descending order
data.groupby(['CompanyType','CompanySize'])['Salary'].mean().sort_values(ascending=False) 
```

In Venture-funded startup , 5,000 to 9,999 employees have the highest average salary among company size.
